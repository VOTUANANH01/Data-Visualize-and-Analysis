### <h3> 2.2.1 Merge bộ dữ liệu </h3>


```python
# link: https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.page
lc_detail=pd.read_csv(path+r"/taxi+_zone_lookup.csv")
#thay thế các giá trị NaN = Unknown
lc_detail=lc_detail.fillna("Unknown")
```

```python
lc_detail
```

```python
def merging(l_df,r_df,*l_key,r_key):
  temp=pd.DataFrame({})
  for a,key in enumerate(l_key):
    #tại l_key đầu tiên thì
    if a == 0:
      #tạo tên gọi thuộc tính mới
      str = key[0:2] +'_'
      #tiến hành merge hai bộ dữ liệu
      df_merge=pd.merge(left=l_df,right=r_df,left_on=key,right_on=r_key)
      #loại bỏ cột thừa
      df_merge=df_merge.drop(r_key,axis=1)
      #lấy ra những tên cột thuộc tính được thêm mới
      name=df_merge.columns.intersection(r_df.columns)
      #gán tên mới cho các cột thuộc tính được thêm
      for i,n in enumerate(name):
        new = str + name[i]
        df_merge.rename(columns={n:new},inplace=True)
      #trả dataframe về 1 biến tạm
      temp=df_merge
    else:
      # tới l_key tiếp theo thực hiện đệ quy lại
      return merging(temp,r_df,key,r_key=r_key)
  #trả về dataframe
  return temp
```

```python
#sau khi merge sẽ sắp xếp lại thứ tự theo cột cuối được chọn
df_merge=merging(df_step3,lc_detail,'PULocationID','DOLocationID',r_key='LocationID')
df_merge
```

### <h3> 2.2.2 Triết xuất data từ hai cột lpep_pickup_datetime và	lpep_dropoff_datetime để có những cột thuộc tính mới </h3>

```python
import datetime as dt
```

```python
# chuyển datatype của lpep_pickup_datetime và	lpep_dropoff_datetime	 dạng date_time
df_merge['lpep_pickup_datetime']= pd.to_datetime(df_merge['lpep_pickup_datetime'])
df_merge ['lpep_dropoff_datetime']= pd.to_datetime(df_merge['lpep_dropoff_datetime'])
```

```python
#quan sát cột 'lpep_pickup_datetime' sau khi thay đổi kiểu dữ liệu
df_merge['lpep_pickup_datetime']
```

```python
#quan sát cột 'lpep_dropoff_datetime' sau khi thay đổi kiểu dữ liệu
df_merge['lpep_dropoff_datetime']
```

```python
#tạo biến thời gian hoàn thành chuyến đi
time_diff=df_merge['lpep_dropoff_datetime']-df_merge['lpep_pickup_datetime']
```

```python
#đưa về đơn vị là phút
for i, time in enumerate(time_diff):
  time_diff[i]=time_diff[i].seconds/60
```

```python
df_time=df_merge.copy()
```

```python
#thêm cột dữ liệu mới
df_time['trip_duration']=time_diff
```

```python
#rút trích data từ các 'lpep_dropoff_datetime' và 'lpep_dropoff_datetime' để lấy ra month (tháng), thứ (được encode:0 ->6 tương đương với bảy ngày), hour(giờ)
for j in ["pickup","dropoff"]:
 df_time[j+"_month"] = df_time['lpep_'+j+'_datetime'].dt.month
 df_time[j+"_weekday"] = df_time['lpep_'+j+'_datetime'].dt.weekday
 df_time[j+"_hour"] = df_time['lpep_'+j+'_datetime'].dt.hour
```

```python
#tạo cột thuộc tính khoảng giờ theo 3 khoảng: [16,20) - rush, [20,6) - overnight, [6,16) - other, được chia cách tính tiền phụ phí của taxi
#https://www1.nyc.gov/site/tlc/passengers/taxi-fare.page
def bin_hour(t):
  if t in range(16,20):
    return "rush"
  if t in range(6,16):
    return "others"
  else:
    return "overnight"

```

```python
df_time["pickup_bin_hours"]=df_time['pickup_hour'].apply(bin_hour)
df_time["dropoff_bin_hours"]=df_time['dropoff_hour'].apply(bin_hour)
```
