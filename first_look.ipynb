```python
#tạo DataFamre df_missing với tỉ lệ khuyết dữ liệu
df_missing=df.isna().melt(value_name="missing")
df_missing=(df_missing.groupby(['variable'])['missing']
                     .value_counts(normalize=True)
                     .rename('percentage')
                     .mul(100)
                     .reset_index())
```

```python
df_missing["missing"] = df_missing["missing"].astype("str")
```

```python
#trực quan dữ liệu bị khuyết
sns.set(rc={'figure.figsize':(15,12)})
plot=sns.barplot(
    data=df_missing,
    x="variable",
    y='percentage',
    hue="missing",
)
fig=plot.get_figure()
plt.xticks(rotation=45)
plt.title("visualizing_missing_data")
plt.show()
fig.savefig(picture_path+r"/visualizing_missing_data.png", dpi=100)
```
